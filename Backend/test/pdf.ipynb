```python
import pdfplumber
import pandas as pd
```

```python
all_rows = dict()
```

```python
with pdfplumber.open("../temp/statement_sample.pdf") as pdf:
    table = pdf.pages[0].extract_tables()[0]

    # headers
    headers = table[0]

    # actual data (single row with \n separated values)
    data_row = table[1]

    # split each column into list
    columns = [col.split("\n") for col in data_row]

    # find max length (number of transactions)
    max_len = max(len(col) for col in columns)

    # normalize columns (fill missing with None)
    normalized_cols = []
    for col in columns:
        if len(col) < max_len:
            col.extend([None] * (max_len - len(col)))
        normalized_cols.append(col)
    
    # transpose → rows
    rows = list(zip(*normalized_cols))

    # convert to list of dicts
    result = []
    for row in rows:
        record = dict(zip(headers, row))
        
        # clean empty strings → None
        for k, v in record.items():
            if v is None or str(v).strip() == "":
                record[k] = None
        
        result.append(record)
```

```python
# preview
for r in result[:5]:
    print(r)
```
